```python
# Convoco librerias.
import pandas as pd
import numpy as np
import locale
from babel.dates import format_date
```

## Cargo el archivo y los divido en dos datasets segun la solapa.

```python
# Especifico el nombre de las hojas que deseas leer.
HECHOS = 'HECHOS'
VICTIMAS = 'VICTIMAS'

# Leo la primera hoja y creo el primer DataFrame.
df_siniestros_hechos = pd.read_excel(r"D:\Users\Arnaldo\Desktop\SISTEMAS\SOYHENRY\CURSO\PROYECTOS\PROYECTO INDIVIDUAL II\RECURSADO\archivos_originales\homicidios.xlsx", sheet_name=HECHOS)

# Lee la segunda hoja y creo el segundo DataFrame.
df_siniestros_victimas = pd.read_excel(r"D:\Users\Arnaldo\Desktop\SISTEMAS\SOYHENRY\CURSO\PROYECTOS\PROYECTO INDIVIDUAL II\RECURSADO\archivos_originales\homicidios.xlsx", sheet_name=VICTIMAS)
```

## Analizo dataset "Hechos".

```python
# Determino dimensiones.
df_siniestros_hechos.shape

# Estudio el detalle de los campos.
df_siniestros_hechos.info()

# Veo el 'Head'
df_siniestros_hechos.head()
```

### Busco y normalizo nulos.

```python
df_siniestros_hechos.replace('sd', np.nan, inplace=True)
nulos1 = df_siniestros_hechos.isna().sum()
nulos1
```

### Busco y elimino datos duplicados.

```python
registros_duplicados = df_siniestros_hechos[df_siniestros_hechos.duplicated()]
registros_duplicados
```

### Inicio transformacion de datos.

```python
df_siniestros_hechos = df_siniestros_hechos.rename(columns=str.lower)
df_siniestros_hechos = df_siniestros_hechos.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# Modifico nombres de columnas "pos x" y "pos y".
df_siniestros_hechos.rename(columns={'pos x': 'longitud', 'pos y': 'latitud'}, inplace=True)

# Convierto las columnas "pos_x" y "pos_y" a tipo de dato float.
df_siniestros_hechos['longitud'] = pd.to_numeric(df_siniestros_hechos['longitud'], errors='coerce').astype(float)
df_siniestros_hechos['latitud'] = pd.to_numeric(df_siniestros_hechos['latitud'], errors='coerce').astype(float)

# Convierto valores nulos a NaN.
df_siniestros_hechos['longitud'] = df_siniestros_hechos['longitud'].replace('.', np.nan)
df_siniestros_hechos['latitud'] = df_siniestros_hechos['latitud'].replace('.', np.nan)
```

```python
# Convierto la columna 'hh' a números, manejar errores y convertir no numéricos a NaN.
df_siniestros_hechos['hh'] = pd.to_numeric(df_siniestros_hechos['hh'], errors='coerce')

# Convierto a 'float64' para admitir NaN.
df_siniestros_hechos['hh'] = df_siniestros_hechos['hh'].astype('float64')

# Convierto a 'Int64' para preservar NaN y tratar los demás valores como enteros.
df_siniestros_hechos['hh'] = df_siniestros_hechos['hh'].astype('Int64')
```

```python
# A fin de facilitar el analisis exploratorio y visualizaciones categorizo las siguientes columnas:

columnas_a_categorizar = ['hh', 'tipo_de_calle', 'comuna', 'victima', 'acusado']

for column in columnas_a_categorizar:
    df_siniestros_hechos[column] = pd.Categorical(df_siniestros_hechos[column])
```

```python
# Establezco el locale en español.
locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')

# COnvierto la columna 'fecha' en tipo datetime.
df_siniestros_hechos['fecha'] = pd.to_datetime(df_siniestros_hechos['fecha'])

# Creo una nueva columna 'dia_semana' con los nombres de los días de la semana en español.
df_siniestros_hechos['dia_semana'] = df_siniestros_hechos['fecha'].apply(lambda x: x.strftime('%A'))

# Muestro el DataFrame con la nueva columna.
df_siniestros_hechos.head(15)
```

#####  Procedo a eliminar columnas que no hacen al analisis.

```python
# Elimino: 'AAAA', 'MM', 'DD', 'HORA', 'LUGAR_DEL_HECHO', 'Calle', 'Altura','Cruce', 'XY (CABA)', 'PARTICIPANTES', 'Dirección Normalizada'.

columnas_eliminar = ['aaaa', 'mm', 'dd', 'hora', 'lugar_del_hecho', 'calle', 'altura','cruce', 'xy (caba)', 'participantes']
df_siniestros_hechos.drop(columnas_eliminar,axis=1,inplace=True)
```

## Analizo dataset "Victimas".

```python
# Determino dimensiones.

df_siniestros_victimas.shape

# Estudio el detalle de los campos.

df_siniestros_victimas.info()

# Mapeo el 'Head'

df_siniestros_victimas.head()
```

### Busco y standarizo nulos.

```python
df_siniestros_victimas.replace('sd', np.nan, inplace=True)
nulos1 = df_siniestros_victimas.isna().sum()
nulos1
```

### Busco y elimino datos duplicados.

```python
# Remplazo 'SD' por NaN en todo el DataFrame.
registros_duplicados = df_siniestros_victimas[df_siniestros_victimas.duplicated()]
df_siniestros_victimas.replace(['SD','sd'], np.nan, inplace=True)

registros_duplicados
```

### Inicio transformacion de datos.

```python
# Renombro columna ID_hecho.
df_siniestros_victimas = df_siniestros_victimas.rename(columns={'ID_hecho': 'ID'})
```

```python
# Llevo a minusculas las columnas y los datos.
df_siniestros_victimas = df_siniestros_victimas.rename(columns=str.lower)
df_siniestros_victimas = df_siniestros_victimas.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
# Convierto la columna 'EDAD' a float y despues a int para preservar Nan.
df_siniestros_victimas['EDAD'] = pd.to_numeric(df_siniestros_victimas['EDAD'], errors='coerce')

# Convierto a 'float64' para admitir NaN.
df_siniestros_victimas['EDAD'] = df_siniestros_victimas['EDAD'].astype('float64')

# Convierto a 'Int64' para preservar NaN y tratar los demás valores como enteros.
df_siniestros_victimas['EDAD'] = df_siniestros_victimas['EDAD'].astype('Int64')
```

```python
# Lista de columnas que se van a categorizar. 
columns_to_categorize = ['ROL', 'SEXO']

for column in columns_to_categorize:
    df_siniestros_victimas[column] = pd.Categorical(df_siniestros_victimas[column])
```

####  Procedo a eliminar columnas que no hacen al analisis.

```python
# Elimino: 'AAAA', 'MM', 'DD'.

columnas_eliminar = ['AAAA', 'MM', 'DD']

df_siniestros_victimas.drop(columnas_eliminar,axis=1,inplace=True)
```
